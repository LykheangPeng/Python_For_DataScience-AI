### 1. Introduction to Python

<b>1.1 What's Python?</b> <br>
Python is a famous and popular programming language well-known by the tech community worldwide. It is a versatile and easy-to-learn language due to its readability. 
Python can be used for various purposes, such as:
- Data Science
- Artificial Intelligence
- Development
- Scientific Computing
- Internet of Things

<b>1.2 Why's Python </b><br>
Python is a high-level, interpreted programming language that supports object-oriented programming. Another reason why Python is so popular is its easy syntax. Additionally, you can simply type and run code without complicated configuration.

<b> 1.3 Keys Tools Installation </b><br>
There are three main tools you need to install for the Python for Data Science & AI course. First, you need the Python package. Second, the code editor of choice for this course is Visual Studio Code. Third, you need Anaconda Navigator, which allows you to create environments to run Python in different versions or for specific purposes, especially useful for data science and AI tasks
